```python
def getValues(n):
    assert isinstance(n, int), "'n' must be an integer"

    items = []
    for i in range(n):
        items.append(float(input("Enter a value: ")))

    if len(items) < 3:
        raise ValueError("Need at least 3 values to exclude the first two")

    return items[2:]


def calcAverage(alist):
    assert isinstance(alist, list), "'alist' must be a list"

    if len(alist) == 0:
        raise ValueError("List is empty after removing first two values")

    return sum(alist) / len(alist)


def main():
    n = int(input("Enter a number of items to be added to the list: "))
    if n < 2:
      raise ValueError("This list must have 3 items")
    alist = getValues(n)
    avg = calcAverage(alist)
    print(f"Average weight: {avg} lbs.")


main()
```

```python
class Greeting:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name

    def display(self, formal):
        if formal:
            print(f"{self.last_name}, {self.first_name}")
        else:
            print(self.first_name)

    def __str__(self):
        return "Class Greeting"


p = Greeting("Zach", "Weiss")
print(p)
p.display(formal=False)
p.display(formal=True)
```

```python
class Student:
  def __init__ (self,name,number):
    self.__name=name
    self.number=number
    self.__courses=[]

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self,newName):
    self.__name=newName

  def __len__(self):
    return len(self.__courses)

  def enroll(self,newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
    else:
      print(f"Student {self.__name} has already enrolled in {newCourse}.")
```

```python
s = Student("John Does", 12345)
print(s.get_name())
print(s.get_courses())
s.enroll('BTE320')
print(len(s))
print(s.get_courses())
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        dx=(self.x-other.x)**2
        dy=(self.y-other.y)**2
        return (dx+dy)**0.5

class Rectangle:
    def __init__(self):
        self.p1 = Point(0,0)
        self.p2 = Point(3,0)
        self.p3 = Point(0,3)
        self.p4 = Point(3,3)

        self.width = self.p1.distance(self.p2)
        self.height = self.p1.distance(self.p3)

    def area(self):
        return self.width * self.height
```

```python
class Employee:
    def __init__(self, name, wage, hours):
        self.name = name
        self._hourly_wage = wage
        self.hours = hours

    def get_wage(self):
        return self._hourly_wage

    def set_wage(self, newWage):
        if newWage <= 0:
            print("Negative wage values not allowed.")
        else:
            self._hourly_wage = newWage

    def calculate_pay(self):
        return self._hourly_wage * self.hours

    def __str__(self):
        return f"Employee: {self.name}\nHourly wage: {self._hourly_wage}"


class Manager(Employee):
    def __init__(self, name, wage, hours, bonus):
        super().__init__(name, wage, hours)
        self.bonus = bonus

    def calculate_pay(self):
        return (1 + self.bonus) * self._hourly_wage * self.hours


employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)

print(employee)
print(manager)
print()

print(f"Employee {employee.name} weekly salary is: {employee.calculate_pay()}$")
print(f"Manager {manager.name} weekly salary is: {manager.calculate_pay()}$ that includes a {manager.bonus * 100}% bonus")
```
