```python
def getValues(n):
    assert isinstance(n, int), "'n' must be an integer"

    items = []
    for i in range(n):
        items.append(float(input("Enter a value: ")))

    if len(items) < 3:
        raise ValueError("Need at least 3 values to exclude the first two")

    return items[2:]


def calcAverage(alist):
    assert isinstance(alist, list), "'alist' must be a list"

    if len(alist) == 0:
        raise ValueError("List is empty after removing first two values")

    return sum(alist) / len(alist)


def main():
    n = int(input("Enter a number of items to be added to the list: "))
    if n < 2:
      raise ValueError("This list must have 3 items")
    alist = getValues(n)
    avg = calcAverage(alist)
    print(f"Average weight: {avg} lbs.")


main()
```

```python
class Student:
  def __init__ (self,name,number):
    self.__name=name
    self.number=number
    self.__courses=[]

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self,newName):
    self.__name=newName

  def __len__(self):
    return len(self.__courses)

  def enroll(self,newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
    else:
      print(f"Student {self.__name} has already enrolled in {newCourse}.")
```

```python
s = Student("John Does", 12345)
print(s.get_name())
print(s.get_courses())
s.enroll('BTE320')
print(len(s))
print(s.get_courses())
```
